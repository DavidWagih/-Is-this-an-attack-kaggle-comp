```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,f1_score,plot_precision_recall_curve,plot_confusion_matrix,plot_roc_curve
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from imblearn.over_sampling import SMOTE 

import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
y=df["Label"]
X=df.drop(["Label"],axis=1)

# drop unused columns
X=X.drop(["ID","Flow ID","Source IP","Destination IP","Timestamp"],axis=1)
```

```python
############ some data cleaning ############ 
```

```python
X["SimillarHTTP"]=X["SimillarHTTP"].apply(lambda x: int(x) if str(x).isnumeric() else 1)
```

```python
X["SimillarHTTP"].value_counts()
```

```python
############## END data cleaning ############ 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y,shuffle=True,random_state=0) 
```

```python
##############  imbalance! solution below ##############
y_train.value_counts()
```

```python
# data scaling over the whole dataset
from sklearn.preprocessing import MinMaxScaler
scaler_all = MinMaxScaler()
X_scaled = scaler_all.fit_transform(X)
```

```python
# mix over + under sampling over train partition only

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

over = RandomOverSampler(sampling_strategy=0.3,random_state=0)
under = RandomUnderSampler(sampling_strategy=0.35,random_state=0)
# over = RandomOverSampler(sampling_strategy=0.55,random_state=0)
# under = RandomUnderSampler(sampling_strategy=1,random_state=0)
#model = LogisticRegression(random_state=0,max_iter=500)
# define pipeline
#pipeline = Pipeline(steps=[('o', over), ('u', under), ('m', model)])
pipeline = Pipeline(steps=[('o', over), ('u', under)])
X_train_resampled, y_train_resampled = pipeline.fit_resample(X_train, y_train)
```

```python
pd.Series(y_train_resampled).value_counts()
```

```python
# mix over + under sampling over the whole dataset - for final train and prediction

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# 0.6 better -> 0.7 worse on precision
over = RandomOverSampler(sampling_strategy=0.55,random_state=0)
under = RandomUnderSampler(sampling_strategy=1,random_state=0)

pipeline = Pipeline(steps=[('o', over), ('u', under)])

X_all_resampled, y_all_resampled = pipeline.fit_resample(X_scaled, y)
```

```python
# before resmapling
print(pd.Series(y).value_counts())
# after resmapling
print(pd.Series(y_all_resampled).value_counts())
```
